```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
```

```python
from catboost import CatBoostClassifier, Pool
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=42, stratify=y_train)
```

```python
model = CatBoostClassifier(iterations=500, depth=6, learning_rate=0.1, random_seed=42)
model.fit(Pool(X_train, y_train), verbose=20)
```

```python
test_preds = model.predict(X_val)
```

```python
accuracy = accuracy_score(y_val, test_preds)
precision = precision_score(y_val, test_preds)
recall = recall_score(y_val, test_preds)
f1 = f1_score(y_val, test_preds)
```

```python
print(f'Точность (Accuracy): {accuracy:.4f}')
print(f'Точность позитивных предсказаний (Precision): {precision:.4f}')
print(f'Полнота (Recall): {recall:.4f}')
print(f'F1-score: {f1:.4f}')
```
