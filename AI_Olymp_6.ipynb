```python
def recall_at_k(y_true, y_scores, k):
    # Убедимся, что y_true и y_scores - это NumPy массивы
    y_true = np.array(y_true)
    y_scores = np.array(y_scores)

    # Сортируем индексы по убыванию вероятностей
    indices = np.argsort(y_scores)[::-1][:k]
    # Выбираем предсказанные классы в топ-k
    top_k_pred = y_true[indices]
    # Считаем положительные классы
    correct_positive_count = np.sum(top_k_pred)
    total_positive_count = np.sum(y_true)

    # Избегаем деления на ноль
    if total_positive_count == 0:
        return 0.0
    return correct_positive_count / total_positive_count
```

```python
import optuna
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import recall_score

def objective(trial):
    solver = trial.suggest_categorical('solver', ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga'])

    # Выбор penalty на основании solver
    if solver == 'liblinear':
        penalty = trial.suggest_categorical('penalty', ['l1', 'l2'])
        l1_ratio = None  # l1_ratio не нужен для liblinear
    elif solver == 'saga':
        penalty = trial.suggest_categorical('penalty', ['elasticnet', 'l1', 'l2'])
        # Для elasticnet требуется l1_ratio
        l1_ratio = trial.suggest_float('l1_ratio', 0, 1) if penalty == 'elasticnet' else None
    else:
        penalty = 'l2'  # Для остальных solvers
        l1_ratio = None  # l1_ratio не нужен для других solvers

    C = trial.suggest_float('C', 1e-5, 1e5, log=True)

    param = {
        'solver': solver,
        'penalty': penalty,
        'C': C,
        'l1_ratio': l1_ratio,  # Добавляем l1_ratio в параметры
        'verbose': False,
        'n_jobs': -1,
        'random_state': 42
    }

    # Обучение модели
    estimator = LogisticRegression(**param)
    estimator.fit(X_train_good, y_train)

    # Предсказания и оценка
    y_scores = estimator.predict(X_test_good)
    recall = recall_at_k(y_val, y_scores, k)

    return recall

# Создание и оптимизация изучения
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=100)

print(study.best_trial)
```
